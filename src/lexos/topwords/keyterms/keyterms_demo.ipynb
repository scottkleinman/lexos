# Simple Keyterms Demo

This notebook shows the basic usage of the `keyterms` module using TextRank as an example.

```python
from lexos.topwords.keyterms.textrank import TextRank, textrank

text = "Natural language processing (NLP) is a subfield of artificial intelligence (AI) that focuses on the interaction between computers and humans through natural language. The ultimate goal of NLP is to enable computers to understand, interpret, and generate human language in a valuable way. Key applications of NLP include machine translation, sentiment analysis, chatbots, and information extraction."
```

## Extract Keyterms from a Plain Text String

```python
# Basic function call
key_terms = textrank(text)
print(key_terms)
```

```python
# Use the TextRank class to get a DataFrame of key terms
TextRank(doc=text).to_df()
```

## Extract Keyterms from a SpaCy `Doc`

```python
# Create a spaCy Doc object
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp(text)

# Extract keyterms from the spaCy Doc object with POS tags
# By default, textrank considers nouns and adjectives as candidate keyterms.
TextRank(doc=doc).to_df()
```
