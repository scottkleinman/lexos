## Import Lexos Modules

```python
from lexos.io.smart import Loader
```

## Load Individual Local Files

```python
data = "../test_data/txt/Austen_Pride.txt"
loader = Loader()
loader.load(data)

#loaded data is accessed using .texts[]
print (loader.texts[0])
```

## Load Lists of Local Files

```python
data = ["../test_data/txt/Austen_Pride.txt",
        "../test_data/docx/Austen_Sense_sm.docx",
        "../test_data/pdf/Austen_Pride_sm.pdf"]

#Loader can handle .txt, .docx, and .pdf file extensions

loader = Loader()
loader.load(data)

text = loader.texts[0]
print(text[0:100])

text = loader.texts[1]
print(text[0:100])

text = loader.texts[2]
print(text[0:100])
```

## Load Local Directories or Zip Files   

Directories or zip files containing files of .txt, .docx, and .pdf extenstions can be loaded.

```python
data = "../test_data/docx"

loader = Loader()
loader.load(data)

data = "../test_data/zip/txt.zip"
loader.load(data)

text = loader.texts[0]
print(text[0:100])

text = loader.texts[1]
print(text[0:100])

text = loader.texts[2]
print(text[0:100])

text = loader.texts[3]
print(text[0:100])

text = loader.texts[4]
print(text[0:100])

text = loader.texts[5]
print(text[0:100])
```

## Load Texts from URL

```python
data = "https://www.gutenberg.org/files/84/84-0.txt"
loader = Loader()
loader.load(data)

print (loader.texts[0])
```
